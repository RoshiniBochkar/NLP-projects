```python
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob, Word
import pandas as pd


nltk.download('stopwords')
nltk.download('wordnet')

stop = set(stopwords.words('english'))

df['Text'] = df['Text'].str.lower()
df['Text'] = df['Text'].str.replace(r'[^\w\s]', ' ', regex=True)
df['Text'] = df['Text'].apply(lambda x: " ".join(word for word in x.split() if word not in stop))
df['Text'] = df['Text'].apply(lambda x: str(TextBlob(x).correct()))
df['Text'] = df['Text'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))

df['Text'].head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

reviews = df.copy()
reviews.dropna(inplace=True)


reviews['Score'].hist(bins=5)
plt.grid(False)
plt.show()


print(reviews.groupby('Score')['Id'].count())
```

```python
import pandas as pd


score_1 = reviews[reviews['Score'] == 1].sample(n=47, replace=True)
score_2 = reviews[reviews['Score'] == 2].sample(n=47, replace=True)
score_3 = reviews[reviews['Score'] == 3].sample(n=47, replace=True)
score_4 = reviews[reviews['Score'] == 4].sample(n=47, replace=True)
score_5 = reviews[reviews['Score'] == 5].sample(n=47, replace=True)
```

```python
reviews_sample=pd.concat([score_1,score_2,score_3,score_4,score_5],axis=0)
reviews_sample.reset_index(drop=True,inplace=True)
(reviews_sample.groupby('Score').count().Id)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

reviews_str = " ".join(reviews_sample["Summary"].astype(str))
wordcloud = WordCloud(background_color="orange").generate(reviews_str)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
negative_reviews=reviews_sample[reviews_sample['Score'].isin([1,2])]
positive_reviews=reviews_sample[reviews_sample['Score'].isin([4,5])]

negative_reviews_str=negative_reviews.Summary.str.cat()
positive_reviews_str=positive_reviews.Summary.str.cat()
```

```python
wordcloud_negative=WordCloud(background_color="black").generate(negative_reviews_str)
wordcloud_positive=WordCloud(background_color="orange").generate(positive_reviews_str)
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(211)
ax1.imshow(wordcloud_negative,interpolation='bilinear')
ax1.axis('off')
ax2=fig.add_subplot(212)
ax2.imshow(wordcloud_positive,interpolation='bilinear')
ax2.axis('off')
ax2.set_title('Reviews with positive Scores',fontsize=20)
plt.show()
```
